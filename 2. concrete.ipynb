```python
%pylab
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv('concrete.csv',sep=',')
data.head(5)
```

```python
#cols = data.columns
cols = ['cement', 'slag', 'ash', 'water', 'superplastic', 'coarseagg','fineagg', 'age', 'strength']
cm = np.corrcoef(data[cols].values.T)
sns.set(font_scale=1.5,style="ticks", color_codes=True)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size':12}, yticklabels=cols, xticklabels=cols)
plt.show()
```
